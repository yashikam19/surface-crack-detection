```python
import os
import random
import numpy as np 
import matplotlib.pyplot as plt
import cv2
```

```python
# Define variables

base_path = "/content/drive/MyDrive/dataset/"
img_size = (224,224,3)
batch_size = 256
```

```python
# function to plot images

def display_img(folder):
    
    images = []
    for filename in os.listdir(folder):
        image = cv2.imread(os.path.join(folder,filename))
        if image is not None:
            images.append(image)
        if len(images)>6:
            break
    
    # plt.figure(figsize=(16,8))    
    for i in range(len(images)):
      plt.imshow(images[i])
      plt.show()
      i+=1
```

```python
# plot cracked images

display_img(base_path + "Positive")
```

```python
# plot non-cracked images

display_img(base_path + "Negative")
```
